## Evaluation metrics

For the validation part of the cluterisation, **the Adjusted rand Index** is going to be used as an evaluation metrics. Given the knowledge of the ground truth class assignments **labels_true** and our clustering algorithm assignments of the same samples **labels_pred**, the adjusted Rand index is a function that measures the similarity of the two assignments, ignoring permutations and with chance normalization.

```python
from sklearn import metrics

labels_true = [0, 0, 0, 1, 1, 1]
labels_pred = [0, 0, 1, 1, 2, 2]

metrics.adjusted_rand_score(labels_true, labels_pred)
```

```python
def see_true_clusterization(feature_one, feature_two, true_labels, xlabel, ylabel):
    fig = plt.figure(figsize=(10,7))
    ax = fig.add_subplot(111)
    scatter = ax.scatter(feature_one, feature_two,
                         c=true_labels, cmap='rainbow', alpha=0.6, s=11)
    ax.set_title('Overview on the true clusterization')
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    plt.colorbar(scatter)
    plt.show()

see_true_clusterization(df_2d10c['Feature1'], df_2d10c['Feature2'], df_2d10c['ClusterIndex'], 'Feature1', 'Feature2')
```

### Hierarchical clustering

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
    
def hierarchycal_clustering(no_of_clusters, condensed_data_matrix, intracluster_metric):
    cluster = AgglomerativeClustering(n_clusters=no_of_clusters, 
                                      affinity='euclidean', linkage=intracluster_metric)
    cluster.fit_predict(condensed_data_matrix)
    return cluster.labels_
```

```python
def visualize_clusterization(feature_one, feature_two, true_labels, predicted_labels, xlabel, ylabel, title):
    adj_rand_index = round(metrics.adjusted_rand_score(true_labels, predicted_labels), 6)
    print('The adjusted Rand Index for ' + title.lower() + ' is: ' + str(adj_rand_index))
    fig = plt.figure(figsize=(10,7))
    ax = fig.add_subplot(111)
    scatter = ax.scatter(feature_one, feature_two,
                         c=predicted_labels, cmap='rainbow', alpha=0.6, s=11)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    plt.colorbar(scatter)
    return plt, adj_rand_index
```

### Density based clustering algorithms

#### DBSCAN

```python
from sklearn.cluster import DBSCAN

def perform_dbscan(data):
    clustering = DBSCAN(min_samples=1).fit(data)
    return clustering.labels_

labels_pred = perform_dbscan(df_2d10c[['Feature1', 'Feature2']])
plot, dbscan_adj_rand_index = visualize_clusterization(df_2d10c['Feature1'], df_2d10c['Feature2'], 
                                                       df_2d10c['ClusterIndex'], labels_pred, 
                                                       'Feature1', 'Feature2', 
                                                       'Clusterization using DBSCAN')
plot.show()
```

### K-means

```python
from sklearn.cluster import KMeans

def perform_kmeans_known_n_clusters(n_clusters, data, feature1, feature2, true_labels, xlabel, ylabel, title):
    kmeans_model = KMeans(n_clusters=n_clusters, init='k-means++', max_iter=100, random_state=0)
    labels_pred = kmeans_model.fit_predict(data)
    plot, kmeans_adj_rand_index = visualize_clusterization(feature1, feature2, true_labels, labels_pred, 
                                   xlabel, ylabel, title)

    # get centers for plot
    centers = kmeans_model.cluster_centers_
    plot.scatter(centers[:, 0], centers[:, 1], c='black', s=30, alpha=0.75)
    plot.show()
    return kmeans_model, kmeans_adj_rand_index

model_kmeans, kmeans_adj_rand_index = perform_kmeans_known_n_clusters(max(list(df_2d10c['ClusterIndex'].unique())) + 1, 
                                      df_2d10c[['Feature1', 'Feature2']], 
                                      df_2d10c['Feature1'], 
                                      df_2d10c['Feature2'], 
                                      df_2d10c['ClusterIndex'], 
                                      'Feature1', 
                                      'Feature2', 
                                      'Clusterization using K-means')
```

### Expectation Maximization (EM)

```python
from sklearn.mixture import GaussianMixture

def perform_gmm(n_clusters, data):
    gmm = GaussianMixture(n_components=n_clusters, random_state=0)
    gmm.fit(data)
    
    labels_pred = gmm.predict(data)
    return labels_pred
 
labels_pred = perform_gmm(max(list(df_2d10c['ClusterIndex'].unique())) + 1, df_2d10c[['Feature1', 'Feature2']].values)
plot, em_adj_rand_index = visualize_clusterization(feature_one = df_2d10c['Feature1'], feature_two = df_2d10c['Feature2'],
                         true_labels = df_2d10c['ClusterIndex'], predicted_labels = labels_pred, 
                         xlabel = 'Feature1', ylabel = 'Feature2', title = 'Clusterization using EM')
```

### Self-organizing maps (SOM)

```python
from minisom import MiniSom    

def perform_som_clusterisation(data, true_labels, title, n_clusters):
    # data normalization
    data = data -  np.mean(data, axis=0)
    data /= np.std(data)

    # Initialization and training
    som_shape = (1, n_clusters)
    som = MiniSom(som_shape[0], som_shape[1], 2, sigma=.3, learning_rate=.5,
                  neighborhood_function='gaussian', random_seed=0)

    som.pca_weights_init(data)
    som.train_batch(data, 1000, verbose=False)

    # each neuron represents a cluster
    winner_coordinates = np.array([som.winner(x) for x in data]).T

    cluster_index = np.ravel_multi_index(winner_coordinates, som_shape)
    adj_rand_index = round(metrics.adjusted_rand_score(true_labels, cluster_index), 6)
    
    print('The adjusted Rand Index for '+ title + ' is: ' + str(adj_rand_index))

    fig = plt.figure(figsize=(10,7))
    # plotting the clusters using the first 2 dimentions of the data
    for c in np.unique(cluster_index):
        plt.scatter(data[cluster_index == c, 0],
                    data[cluster_index == c,1], label='cluster='+str(c), alpha=.7, s=13)

    # plotting centroids
    for centroid in som.get_weights():
        plt.scatter(centroid[:, 0], centroid[:, 1], marker='x', 
                    s=80, linewidths=35, color='k', label='centroid')

    plt.title(title)
    plt.legend();
    
    return adj_rand_index
    
data = df_2d10c[['Feature1', 'Feature2']].values
true_labels = df_2d10c['ClusterIndex']
title = 'SOM clusterization'
n_clusters = max(list(df_2d10c['ClusterIndex'].unique())) + 1

som_adj_rand_index = perform_som_clusterisation(data, true_labels, title, n_clusters)
```
